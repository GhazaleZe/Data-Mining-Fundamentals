```python
breaks=jenkspy.jenks_breaks(df['age'], nb_class=6)
print(breaks)
df['age']=pd.cut(df['age'],bins=breaks,labels=[1,2,3,4,5,6])

breaks1=jenkspy.jenks_breaks(df['serum cholestoral '], nb_class=3)
print(breaks1)
df['serum cholestoral ']=pd.cut(df['serum cholestoral '],bins=breaks1,labels=[1,2,3])

breaks2=jenkspy.jenks_breaks(df['maximum heart rate achieved'], nb_class=3)
print(breaks2)
df['maximum heart rate achieved']=pd.cut(df['maximum heart rate achieved'],bins=breaks2,labels=[1,2,3])
df
```

```python
df=pd.get_dummies(df,columns=['age','serum cholestoral ','maximum heart rate achieved','chest pain type','target','gender','exercise induced angina'])
df.head()
```

```python
df.replace(1,True,inplace=True)
df.replace(2,True,inplace=True)
df.replace(0,False,inplace=True)
df
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
frq_items=apriori(df, min_support=0.2,use_colnames=True)
frq_items
```

```python
rules = association_rules(frq_items, metric='lift')
rules = pd.concat([rules[rules.consequents == frozenset({'target_1'})],
                   rules[rules.consequents == frozenset({'target_0'})]], ignore_index=True)
rules.sort_values(by=['consequents','confidence','lift','support'], inplace=True, ascending=False)
rules
```
